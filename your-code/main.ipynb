```python

outliers_list = []
for i in outliers:
    low = outliers[i].mean() -2*outliers[i].std();
    high = outliers[i].mean() +2*outliers[i].std();
    outliers_row = []
    for j in range(len(outliers[i])):
        if outliers[i][j]<low or outliers[i][j]>high:
            outliers_row.append(outliers[i][j])
        else:
            outliers_row.append(np.NaN)
    outliers_list.append(outliers_row)
    


outliers_df = pd.DataFrame(outliers_list)
outliers_df = outliers_df.transpose()
outliers_df.columns=['Reputation', 'Views', 'UpVotes', 'Score', 'ViewCount', 'AnswerCount', 'CommentCount', 'FavoriteCount']
outliers_df = outliers_df.dropna(how='all')
```
